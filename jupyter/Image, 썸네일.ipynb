```python
import requests
```

```python
image_url = 'https://images.pexels.com/photos/39517/rose-flower-blossom-bloom-39517.jpeg?auto=compress&cs=tinysrgb&h=650&w=940'
```

```python
image_binary = requests.get(image_url).content
with open('flower.jpg', 'wb') as f:
    f.write(image_binary)
```

```python
from PIL import Image  # Pillow
flower_image = Image.open('flower.jpg')
```

```python
flower_image
```

```python
from pilkit.processors import Thumbnail
processor = Thumbnail(width=300, height=300)
```

```python
thumb_image = processor.process(flower_image)
```

```python
thumb_image
```

```python
thumb_image.save("thumb-300x300.png")
thumb_image.mode = 'RGB'
thumb_image.save("thumb-300x300x20.jpg", quality=20)
thumb_image.save("thumb-300x300x80.jpg", quality=80)
```
